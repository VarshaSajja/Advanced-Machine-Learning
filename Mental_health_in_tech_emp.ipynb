# Abstract

Mental health in the workplace (specifically in tech) has become an important issue that was brought to light in recent years. Researchers are trying to determine what causes mental health issues and if there is a way to improve workplace conditions to help combat mental health.

```python
# Different combinations between the two Questions
# Q1: 'Do you currently have a mental health disorder?'
# Q2: 'Have you had a mental health disorder in the past?'
# (Q1,Q2) = JointResult (Highest answer - Order: Y M N)
# (y,y) = y
# (y,m) = y
# (y,n) = y
# (m,y) = y
# (m,m) = m
# (m,n) = m
# (n,y) = y
# (n,m) = m
# (n,n) = n

# To simplify Code, Change Column Names
tempdf['Q1']= tempdf['Have you had a mental health disorder in the past?']
tempdf['Q2'] = tempdf['Do you currently have a mental health disorder?']

def mentalHealth(tempdf):
    if ((tempdf['Q1'] == 'Yes' and tempdf['Q2'] == 'Yes') or
        (tempdf['Q1'] == 'Yes' and tempdf['Q2'] == 'Maybe') or
       (tempdf['Q1'] == 'Yes' and tempdf['Q2'] == 'No') or
       (tempdf['Q1'] == 'Maybe' and tempdf['Q2'] == 'Yes') or
       (tempdf['Q1'] == 'No' and tempdf['Q2'] == 'Yes')):
        return 'Yes'
    elif ((tempdf['Q1'] == 'Maybe' and tempdf['Q2'] == 'Maybe') or
        (tempdf['Q1'] == 'Maybe' and tempdf['Q2'] == 'No') or
       (tempdf['Q1'] == 'No' and tempdf['Q2'] == 'Maybe')):
      return 'Maybe'
    else:
        return 'No'

tempdf['Had or Has Mental Health Disorder'] = tempdf.apply(mentalHealth, axis=1)                              
```

```python
# Create Temp dataset for this research question:
#   Add the Has or Had Mental Health Disorder Variable
#   Add the Continent Variable to one data set and the keep the country in the other
#   Add the summarized work position variable to give that variable more meaning

cols_to_keep = ['What is your age?', 'What is your gender?',
       'What country do you work in?',
       'Which of the following best describes your work position?',
       'Do you work remotely?','Have you had a mental health disorder in the past?',
       'Do you currently have a mental health disorder?']
r1_df = df[cols_to_keep]

######

# To simplify Code, Change Column Names
q1 = 'Have you had a mental health disorder in the past?'
q2 = 'Do you currently have a mental health disorder?'

def mentalHealth(r1_df):
    if ((r1_df[q1] == 'Yes' and r1_df[q2]== 'Yes') or
        (r1_df[q1] == 'Yes' and r1_df[q2] == 'Maybe') or
       (r1_df[q1] == 'Yes' and r1_df[q2] == 'No') or
       (r1_df[q1] == 'Maybe' and r1_df[q2] == 'Yes') or
       (r1_df[q1] == 'No' and r1_df[q2] == 'Yes')):
        return 1
    elif ((r1_df[q1] == 'Maybe' and r1_df[q2] == 'Maybe') or
        (r1_df[q1] == 'Maybe' and r1_df[q2] == 'No') or
       (r1_df[q1] == 'No' and r1_df[q2] == 'Maybe')):
      return 0
    else:
        return 0

r1_df['Had or Has Mental Health Disorder'] = r1_df.apply(mentalHealth, axis=1)

#######

q3 = 'What country do you work in?'

def continent(r1_df):
  
  USA = ['United States of America']
  Canada = ['Canada']
  Mexico = ['Mexico']
  Europe = ['United Kingdom', 'Germany', 'Netherlands', 'Czech Republic', 'Lithuania',
            'France', 'Poland', 'Belgium', 'Denmark','Sweden', 'Russia', 'Spain','Romania',
            'Switzerland', 'Norway','Ireland', 'Italy','Finland', 'Turkey','Bulgaria', 
            'Slovakia', 'Austria','Greece', 'Hungary', 'Israel', 'Serbia', 'Estonia',
            'Bosnia and Herzegovina']
  Australia = ['Australia', 'New Zealand']
  Asia = ['Japan','India','United Arab Emirates','Pakistan','Afghanistan',
          'Vietnam','Bangladesh','Iran','Iran']
  C_S_America = ['Venezuela', 'Brazil','Argentina', 'Colombia', 'Costa Rica',
               'Ecuador','Chile', 'Guatemala']
  Africa = ['South Africa']

  if r1_df[q3] in USA:
    return 'USA'
  elif r1_df[q3] in Canada:
    return 'Canada'
  elif r1_df[q3] in Mexico:
    return 'Mexico'
  elif r1_df[q3] in Europe:
    return 'Europe'
  elif r1_df[q3] in Australia:
    return 'Australia'
  elif r1_df[q3] in Asia:
    return 'Asia'
  elif r1_df[q3] in C_S_America:
    return 'Central/South America'
  elif r1_df[q3] in Africa:
    return 'Africa'
  else:
    return 'Other'

r1_df['Continent'] = r1_df.apply(continent, axis=1)

######


def positionGeneral(r1_df):
  
  q4 = 'Which of the following best describes your work position?'
  
  BE_Dev = ['Back-End Developer','Back-end Developer|DevOps/SysAdmin','DevOps/SysAdmin']
  FE_Dev = ['Front-End Developer','Front-end Developer|Designer','Designer','Supervisor/Team Lead|Back-end Developer|Front-end Developer']
  Both_Dev = ['Back-end Developer|Front-end Developer','Front-end Developer|Back-end Developer','DevOps/SysAdmin|Back-end Developer|Front-end Developer','Front-end Developer|Back-end Developer|DevOps/SysAdmin']
  Supervisor = ['Supervisor/Team Lead','Executive Leadership','One-person shop','Supervisor/Team Lead|Back-end Developer']
  
  if r1_df[q4] in BE_Dev:
    return 'Back-end Developer'
  elif r1_df[q4] in FE_Dev:
    return 'Front-end Developer'
  elif r1_df[q4] in Both_Dev:
    return 'Back-end and Front-end Developer'
  elif r1_df[q4] in Supervisor:
    return 'Leadership'
  else:
    return 'Other'

r1_df['Position_General'] = r1_df.apply(positionGeneral, axis=1)

######
```

#### Preprocessing - Part 1

```python
# Different combinations between the two Questions
# Q1: 'Do you have a family history of mental illness?'
# Q2: 'Have you had a mental health disorder in the past?'
# (Q1,Q2) = JointResult (Highest answer - Order: Yes    Maybe/I Don't know    No)
# (y,y) = y
# (y,m) = y
# (y,n) = y
# (m,y) = y
# (m,m) = m
# (m,n) = m
# (n,y) = y
# (n,m) = m
# (n,n) = n

# To simplify Code, Change Column Names
tempdf['Q1']= tempdf['Do you have a family history of mental illness?']
tempdf['Q2'] = tempdf['Have you had a mental health disorder in the past?']

def mentalHealth(tempdf):
    if ((tempdf['Q1'] == 'Yes' and tempdf['Q2'] == 'Yes') or
        (tempdf['Q1'] == 'Yes' and tempdf['Q2'] == 'Maybe') or
       (tempdf['Q1'] == 'Yes' and tempdf['Q2'] == 'No') or
       (tempdf['Q1'] == "I don't know" and tempdf['Q2'] == 'Yes') or
       (tempdf['Q1'] == 'No' and tempdf['Q2'] == 'Yes')):
        return 'Yes'
    elif ((tempdf['Q1'] == "I don't know" and tempdf['Q2'] == 'Maybe') or
        (tempdf['Q1'] == "I don't know" and tempdf['Q2'] == 'No') or
       (tempdf['Q1'] == 'No' and tempdf['Q2'] == 'Maybe')):
      return 'Maybe'
    else:
        return 'No'

tempdf['Had history of Mental illness'] = tempdf.apply(mentalHealth, axis=1)                              
```

```python
def print_metrics(true, pred):    
    print('\nCONFUSION MATRIX')
    tn, fp, fn, tp = metrics.confusion_matrix(true, pred).ravel()
    print(metrics.confusion_matrix(true, pred))
    precision_score = metrics.precision_score(true, pred)
    recall_score = metrics.recall_score(true, pred)
    specificity = tn / (tn+fp)
    fpr = fp / (fp + tn)
    f1_score = metrics.f1_score(true, pred)
    print("\nMetrics on test data")
    print('Precision Score :', round(precision_score,2))
    print('Recall Score :', round(recall_score,2))
    print('Specificity :', round(specificity,2))
    print('False Positive Rate :', round(fpr,2))
    print('F1 Score :', round(f1_score,2))
```

```python
from sklearn import metrics

# helper function to fit and evaluate a model
def evaluation_metrics(model, model_name, x_train=x_train, x_test=x_test, y_train=y_train, y_test=y_test):
    model.fit(x_train, y_train)

    y_pred_train = model.predict(x_train)
    print(f'Training Accuracy: {metrics.accuracy_score(y_train, y_pred_train)}')
  
    y_pred_test = model.predict(x_test)
    print(f'Testing Accuracy: {metrics.accuracy_score(y_test, y_pred_test)}')

    # create a confusion matrix
    confusion_matrix = metrics.confusion_matrix(y_test, y_pred_test)
    ax = sns.heatmap(confusion_matrix, annot=True, square=False, fmt='g', cbar=False)
    bot, top = ax.get_ylim()
    ax.set_ylim(bot + .5, top - .5)
    plt.title(f"{model_name}", fontsize = 15)
    plt.ylabel('Expected')
    plt.xlabel('Predicted')
    plt.show()

    # calculate additional metrics for comparing models
    true_neg = confusion_matrix[0][0]
    true_pos = confusion_matrix[1][1]
    false_pos = confusion_matrix[0][1]
    false_neg = confusion_matrix[1][0]
    
    recall = metrics.recall_score(y_test, y_pred_test, pos_label='Yes')
    specificity = true_neg / (true_neg + false_pos)
    precision = true_pos / (true_pos + false_pos)
    balanced_accuracy = metrics.balanced_accuracy_score(y_test, y_pred_test)
    f1_score = metrics.f1_score(y_test, y_pred_test, pos_label='Yes')

    print(f'Recall: {recall}')
    print(f'Specificity: {specificity}')
    print(f'Precision: {precision}')
    print(f'Balanced Accuracy: {balanced_accuracy}')
    print(f'F1 Score: {f1_score}')

    return model
```
